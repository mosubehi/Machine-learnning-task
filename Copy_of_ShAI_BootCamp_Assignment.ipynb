```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/Salaries.csv')
df.head()
```

```python
df.columns
```

```python
#---------- Identify the number of rows and columns in the dataset
rows=len(df.axes[0])
columns=len(df.axes[1])
print("the number of rows = ",rows)
print("the number of columns = ",columns)
#---determine the data types of each column
df.dtypes
#------check for missing values in each column.
df.isnull()
```

```python
#------mean, median, mode, minimum, and maximum salary
df['TotalPay'].mean()
df['TotalPay'].median()
df['TotalPay'].mode()
mn=df['TotalPay'].min()
mx=df['TotalPay'].max()
print("the minimum ",mn)
print("the maximum ",mx)
#------determine the range of salaries
#avg=df['BasePay'].sum()/df['BasePay'].count() this code for finding thr average.....sorry...
print("the range is ",mx-mn)
#-----find the standard deviation.
sd=np.std(df['TotalPay'])
print("the standard deviation ",sd)
```

```python
#I opened the salaries file Status and Notes columns are empty so we can clean the data set by deleting them
df=df.drop(columns=["Notes","Status"])
#Benefits column has missing values so we can fill them by zeros
df["Benefits"].fillna(0,inplace=True)
#the data set is huge so i can't determin if we have a duplicate rows so i will apply this line of code to ensure that the duplicates removed
df.drop_duplicates(inplace=True)

```

```python
#df.head()
columns=list(df.columns)
print(columns)
```

```python
df.hist(column='TotalPay',color='orange')
```

```python
group_=df.groupby(['JobTitle'])
group_.head()
stats=group_.agg({'Year':['mean','sum']})
print(stats)
avg=group_.agg({'TotalPay':['mean']})
print(avg)
```

```python
import seaborn as sns
corr_=df['TotalPay'].corr(df['TotalPayBenefits'])
graph_=sns.scatterplot(x='TotalPay',y='TotalPayBenefits',data=df)
graph_.set_title("salary vs total")
```
